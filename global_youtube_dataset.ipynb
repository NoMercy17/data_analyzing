```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("Global YouTube Statistics.csv", encoding='latin1')

df.head(3)
```

```python
df.info()
df.describe()
```

```python
df.isna().sum()
df.duplicated().sum()
```

```python
df.columns
```

```python
pd.set_option("display.float_format", "{:.0f}".format)
category_subscribers_mean = df.groupby('category')['subscribers'].mean()
category_subscribers_mean.sort_values(ascending=False, inplace=True)
```

```python
category_subscribers_mean.plot(kind = 'bar') # see the categ information
```

```python
channel_type_counts = df["channel_type"].value_counts()
channel_type_counts
```

```python
channel_type_counts.plot(kind = 'pie')
plt.xlabel('Channel Type Distributino')
```

```python
plt.scatter(df['subscribers'], df['video views'])
```

```python
country_subscribers_total = df.groupby('Country')['subscribers'].sum()              
country_subscribers_total
```

```python
country_subscribers_total.head(10).plot(kind = 'bar', grid=True)
plt.title('Total Subscribers by Country')
plt.xlabel("Country")
plt.ylabel("Total subscribers")
plt.xticks(rotation = 60)
plt.show()
```

```python
# we search for how many subscribers we have in a year
df.groupby('created_year')['subscribers'].mean().sort_values(ascending=True)
```

```python
# we calc the median video views by channel type
df.groupby('channel_type')['video_views_rank'].median().sort_values(ascending=True)
```

```python
# the nr of subscribers in the the last 30 months from every country
df.groupby('Country')['subscribers_for_last_30_days'].max().sort_values(ascending=True)
```

```python
df.groupby('created_month')['subscribers'].sum()
```

```python
# total uploads based on channel
df.groupby('channel_type')['uploads'].sum()
```

```python
df.groupby(['Latitude','Longitude']).size()
df.columns
```

```python
sns.boxplot(x='channel_type', y = 'highest_monthly_earnings', data=df)
plt.xlabel('channel_type')
plt.ylabel('highest_monthly_earning')
plt.xticks(rotation = 60)
plt.show()
```

```python
df.groupby(['category', 'Country'])['subscribers'].sum().unstack().plot(kind = 'bar', stacked=True)
plt.xlabel("category")
plt.ylabel("total subscribers")
plt.xticks(rotation = 90)
plt.legend(title = "Country", bbox_to_anchor =(1, 1), loc = 'upper left')
plt.show()
```
